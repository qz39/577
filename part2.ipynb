```python
class Polynomial(object):
    def __init__(self, coef):
        self.coef = coef 
        
    def degree(self):
        degree = len(self.coef) - 1 
        return degree

    def __repr__(self):
        degree = self.degree()
        string = ""
        for i, a in enumerate(self.coef):
            string += f"{a}x^{degree - i} + "
        return string[:-2] 

    def evaluate(self, x):
        degree = self.degree()
        value = 0
        for i, a in enumerate(self.coef):
            value += a * x**(degree - i)
        return value


p = Polynomial([6, 0, -7, 5])

print(p)
print("p(x) = x^3 - 7x + 5")
print(f"p(x) = {p}")
print(f"p(3) = {p.evaluate(2)}")
```

```python
class Polynomial(object):
    # Call the __init__ function which initializes the attributes of the object 
    def __init__(self, coef):
        self.coef = coef # Does not need to have the same name
    # Define a degree method
    def degree(self):
        degree = len(self.coef) - 1 
        return degree

    def __repr__(self):
        degree = self.degree()
        string = ""
        for i, a in enumerate(self.coef):
            string += f"{a}x^{degree - i} + "
        return string[:-2] 

    def __call__(self, x):
        degree = self.degree()
        value = 0
        for i, a in enumerate(self.coef):
            value += a * x**(degree - i)
        return value

p = Polynomial([6, 0, -7, 5])

print(p)
print("p(x) = 6x^3 - 7x + 5")
print(f"p(x) = {p}")
print(f"p(3) = {p(3)}")
```
